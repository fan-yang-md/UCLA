```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Poisson distributions
lambda_1 = 2
lambda_2 = 8
lambda_sum = 10
sample_size = 10**6

# Generate samples for X ∼ Poisson(λ1) and Y ∼ Poisson(λ2)
X_samples = np.random.poisson(lambda_1, sample_size)
Y_samples = np.random.poisson(lambda_2, sample_size)

# Compute Z by summing X and Y
Z_constructed = X_samples + Y_samples

# Generate samples directly from Z ∼ Poisson(λ1 + λ2)
Z_direct = np.random.poisson(lambda_sum, sample_size)

# Plot histograms of both Z_constructed and Z_direct
plt.figure(figsize=(12, 6))
plt.hist(Z_constructed, bins=50, label='Constructed Z (Z= X + Y)', color='green', density=True)
plt.hist(Z_direct, bins=50, label='Directly Sampled Z', color='red', density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Comparison of Constructed Z (Z = X + Y) vs Directly Sampled Z')
plt.show()
```

```python
# Parameters for Poisson distributions
lambda_1 = 2
lambda_2 = 8
lambda_sum = 10
sample_size = 10**6

# Parameters for the conditional distribution
t = 20
p = 1/5

# Generate samples for X ∼ Poisson(λ1) and Y ∼ Poisson(λ2)
X_samples = np.random.poisson(lambda_1, sample_size)
Y_samples = np.random.poisson(lambda_2, sample_size)

# Compute Z by summing X and Y
Z_constructed = X_samples + Y_samples

# Filter Z_constructed samples where X + Y = t
stratified_X = X_samples[Z_constructed == t]

# Generate samples directly from the Binomial distribution Binomial(t, p)
binomial_direct = np.random.binomial(t, p, size=len(stratified_X))

# Plot histograms of both filtered X given X + Y = t and directly sampled Binomial distribution
plt.figure(figsize=(12, 6))
plt.hist(stratified_X, bins=20, label='Stratified X', color='green', density=True)
plt.hist(binomial_direct, bins=20, label='Directly Sampled Binomial', color='red', density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Comparison of Stratified X vs Directly Sampled Binomial')
plt.show()
```

```python


# Parameters for the bi-variate standard normal distribution
mean = [0, 0]  # Mean vector
covariance = [[1, 0], [0, 1]]  # Covariance matrix
sample_size = 1000  # Number of samples to draw

# Sample from the bi-variate standard normal distribution
samples = np.random.multivariate_normal(mean, covariance, sample_size)

# Separate the samples into X and Y for plotting
X_values= samples[:, 0]
Y_values = samples[:, 1]

# Scatter plot of the samples
plt.figure(figsize=(8, 8))
plt.scatter(X_values, Y_values, marker='.')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bivariate Standard Normal Distribution')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
#Count samples where X >= 0
count_X = np.sum(X_values >= 0)
P_X = count_X / sample_size

#Count samples where X >= 0 and Y >= 0
count_XY = np.sum((X_values >= 0) & (Y_values >= 0))

#Count samples where Y >= 0
count_Y = np.sum(Y_values >= 0)

#Calculate conditional probability P(X >= 0 | Y >= 0)
P_XY = count_XY / count_Y if count_Y > 0 else 0

print("P(X ≥ 0):", P_X)
print("P(X ≥ 0 | Y ≥ 0):", P_XY)
```
